### Now let us see which clubs have the best players

Creating a DataFrame for the total ratings for all the players

```
best_player = {}
for club in df['Club'].unique():
    overall_rating = df['Overall'][df['Club'] == club].sum()
    best_player[club] = overall_rating
df_bsp = pd.DataFrame.from_dict(best_player , orient = 'index' , 
                                                 columns = ['overall'])
df_bsp['club'] = df_bsp.index
df_bsp = df_bsp.sort_values(by = 'overall' , ascending =  False)
```
